```python
import pandas as pd

cols = ['path', 'province', 'speaker', 'duration']
data = pd.read_csv('VIN27/combined_audio_data.csv', names=cols)
data.head()
```

```python
# data > 3s
data_long = data[data['duration'] >= 3]
data_long['duration'].sum() / 3600
```

```python
# data < 2s
data_2s = data[data['duration'] < 2]
data_2s['duration'].sum() / 3600
```

```python
# data 2 to 3s
data_2to3 = data[(data['duration'] > 2) & (data['duration'] < 3)]
# sorted_2to3 = data_2to3.sort_values(by='duration', ascending=True)
# sorted_2to3.head(1000)

def sample_and_concat_by_duration(df, x):
  """
  Samples data from a DataFrame based on duration ranges and concatenates the results.

  Args:
    df: Pandas DataFrame with a 'duration' column.
    x: Number of samples to take from each duration range.

  Returns:
    A DataFrame containing the concatenated samples from all duration ranges.
  """

  sampled_data_list = []
  for i in range(20, 30):
    lower_bound = i / 10
    upper_bound = (i + 1) / 10
    
    # Filter data for the current duration range
    df_range = df[(df['duration'] >= lower_bound) & (df['duration'] < upper_bound)]
    
    # Sample x samples from the filtered data, if available
    sampled_data_list.append(df_range.sample(n=x, replace=True) if len(df_range) >= x else df_range)
  
  # Concatenate all sampled DataFrames into one
  return pd.concat(sampled_data_list, ignore_index=True)

# Example usage:
final_2to3 = sample_and_concat_by_duration(data_2to3, x=10000)
print(final_2to3['duration'].sum() / 3600)
print(data_2to3['duration'].sum() / 3600)
```
